### 📚 모범답안: 배경지식 퀴즈 3. 지수함수 / 로그함수 복습

1. **\(y = 2^x\) 의 \(x = 0, 1, 2, 3, 4\) 에서의 \(y\) 값:**
   - \(x = 0\) : \(y = 1\)
   - \(x = 1\) : \(y = 2\)
   - \(x = 2\) : \(y = 4\)
   - \(x = 3\) : \(y = 8\)
   - \(x = 4\) : \(y = 16\)

   그래프를 그리면, \(y = 2^x\) 는 빠르게 증가하는 곡선 형태를 가집니다.

2. **\(y = \log_2(x)\) 의 \(x = 1, 2, 4, 8, 16\) 에서의 \(y\) 값:**
   - \(x = 1\) : \(y = 0\)
   - \(x = 2\) : \(y = 1\)
   - \(x = 4\) : \(y = 2\)
   - \(x = 8\) : \(y = 3\)
   - \(x = 16\) : \(y = 4\)

   그래프를 그리면, \(y = \log_2(x)\) 는 \(x\)가 커질수록 느리게 증가하는 곡선 형태를 가집니다.

3. **그래프의 시작점:**
   - **\(y = 2^x\)** 그래프는 \((0, 1)\)에서 시작합니다.
   - **\(y = \log_2(x)\)** 그래프는 \((1, 0)\)에서 시작합니다.

4. **그래프 대칭:**
   - \(y = 2^x\) 그래프를 \(y = x\) 기준으로 대칭시키면 \(y = \log_2(x)\) 그래프가 됩니다. 이는 지수 함수와 로그 함수가 서로 역함수임을 보여줍니다.

5. **증가 속도 비교:**
   - **\(y = \log_2(x)\) vs. \(y = x\):** \(y = \log_2(x)\)가 더 느리게 증가합니다.
   - **\(y = 2^x\) vs. \(y = x\):** \(y = x\)가 더 느리게 증가합니다. \(y = 2^x\)는 매우 빠르게 증가합니다.

> **정리:** 데이터 개수가 늘어날 때 실행 횟수가 적을수록 좋습니다. 따라서 로그 함수가 선형 함수보다 더 효율적이며, 지수 함수는 비효율적입니다.

### 📚 모범답안: 배경지식 퀴즈 4. 시간복잡도 기초

**작업 순위 (오래 걸리는 순서대로):**

1. **2번: 100개의 파일을 읽고 각 파일 모든 단어를 알파벳순으로 정렬한 후 출력하기.**
   - 시간복잡도: \(O(n \log n)\) — 각 파일의 단어를 정렬해야 하므로 가장 시간이 많이 걸립니다.

2. **3번: 100개의 파일을 읽고 각 파일의 단어 수를 세어 출력하기.**

3. **1번: 100개의 파일을 하나씩 읽고 각 제목을 출력하기.**
   - 시간복잡도: \(O(1)\) — 파일 제목만 읽고 출력하므로 가장 적은 시간이 걸립니다.

> **정리:** 파일을 읽고 정렬하는 작업이 가장 오래 걸리고, 제목만 출력하는 작업이 가장 짧은 시간이 걸립니다.
