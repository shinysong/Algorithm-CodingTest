- **큐 (FIFO):** 먼저 추가된 항목이 먼저 제거됩니다. (버스 정류장 줄서기, 엘리베이터 타기)

### 문제2 모범답안

```python
# 딕셔너리(해시 테이블) 생성
my_dict = {}
my_dict['영수'] = 25
my_dict['민수'] = 30
my_dict['철수'] = 29
print(my_dict)  # {'영수': 25, '민수': 30, '철수': 29}

# 영수의 나이를 print해 주세요.
print(my_dict['영수'])  # 25

# 철수의 생일입니다. 철수의 나이를 30으로 업데이트해 주세요.
my_dict['철수'] = 30

# 민수가 민준으로 개명했습니다. 민준의 이름을 업데이트해 주세요.
my_dict['민준'] = my_dict.pop('민수')

# 새 순서쌍 넣기
# 24세인 진수를 추가해 주세요.
my_dict['진수'] = 24

# 특정 키 가진 순서쌍 삭제
# 영수의 정보를 딕셔너리에서 삭제해 주세요.
del my_dict['영수']

# 특정한 값(value) 가진 키-값 쌍 삭제
# 나이가 25세 이하인 사람의 정보를 모두 삭제해 주세요.
my_dict = {key: value for key, value in my_dict.items() if value > 25}

# 최종 딕셔너리 출력
print(my_dict)  # {'철수': 30, '민준': 30}
```
