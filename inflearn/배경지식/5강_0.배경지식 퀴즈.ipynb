### 2번 문제 예시 답안: 규칙을 파악하여 점화식을 만들기

**수열:**  
\(2, 5, 10, 17, 26, ...\)

**점화식 규칙:**  
이 수열은 \(n\)번째 항이 \(n-1\)번째 항에 \(2n - 1\)을 더한 형태입니다.

**점화식:**  
- \( a(1) = 2 \)
- \( a(n) = a(n-1) + (2n - 1) \) for \( n \geq 2 \)

**재귀 함수 표현:**

```python
def sequence_rule(n):
    if n == 1:
        return 2
    return sequence_rule(n-1) + (2 * n - 1)

# 예시로, sequence_rule(5)의 값을 출력해 보세요.
print(sequence_rule(5))  # 예상 출력: 26
```





### 3번 문제 예시 답안: 규칙을 파악하여 점화식을 만들기

**수열:**  
\(3, 6, 13, 28, 59, ...\)

**점화식 규칙:**  
이 수열은 \(n\)번째 항이 \(n-1\)번째 항에 \(2(n-1) + 1\)을 곱한 후, 추가적으로 \(1\)을 더한 형태입니다.

**점화식:**  
- \( a(1) = 3 \)
- \( a(n) = a(n-1) \times 2 + (n - 1) + 1 \) for \( n \geq 2 \)

**재귀 함수 표현:**

```python
def complex_sequence(n):
    if n == 1:
        return 3
    return complex_sequence(n-1) * 2 + (n - 1) + 1

# 예시로, complex_sequence(5)의 값을 출력해 보세요.
print(complex_sequence(5))  # 예상 출력: 59
```
