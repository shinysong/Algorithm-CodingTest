### 배경지식 퀴즈 1: 트리 기본 용어

다음 중 트리 구조에 대한 설명으로 올바른 것은 무엇인가요?

1. 트리의 모든 노드는 반드시 두 개의 자식 노드를 가져야 한다.
2. 트리에는 루트 노드가 하나 있고, 루트에서 각 노드까지의 경로는 유일하다.
3. 트리에서는 사이클이 존재할 수 있다.
4. 트리의 모든 노드는 서로 연결되어 있지 않아도 된다.

### 배경지식 퀴즈 2: 트리와 그래프의 기본 개념

다음 중 트리와 그래프에서 공통적으로 사용되는 용어가 아닌 것은 무엇인가요?

1. 노드(Node)
2. 간선(Edge)
3. 루트(Root)
4. 사이클(Cycle)

#### 배경지식 문제 답안
1번 답안 :
> **정답:** 2. 트리에는 루트 노드가 하나 있고, 루트에서 각 노드까지의 경로는 유일하다.  
> **설명:** 트리는 하나의 루트 노드에서 시작하며, 각 노드는 부모와의 유일한 경로로 연결됩니다. 트리에는 사이클이 존재하지 않으며, 모든 노드는 연결되어 있어야 합니다.

2번 답안 :  4. 사이클(Cycle)  -  트리는 사이클이 없는 구조이고, 그래프는 사이클이 존재할 수 있습니다.

3번 답안 :
  가운데 트리의 모든 leaf nodes : F, E, H, G

  이진트리 판단 - 이진트리의 정의는 각 노드가 최대 두 개의 자식 노드를 가질 수 있는 트리입니다. 이미지를 살펴보면 다음과 같은 결론을 내릴 수 있습니다:

  - **첫 번째 트리**는 각 노드가 두 개 이하의 자식 노드를 가지므로 **이진트리**입니다.
  - **두 번째 트리**도 각 노드가 두 개 이하의 자식 노드를 가지므로 **이진트리**입니다.
  - **세 번째 트리**는 **노드 B**가 세 개의 자식 노드 (F, D, H)를 가지고 있으므로 **이진트리가 아닙니다**.

  따라서, **세 번째 트리**는 이진트리가 아닙니다.
